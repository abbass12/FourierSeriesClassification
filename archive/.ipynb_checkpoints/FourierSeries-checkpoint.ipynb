## Step Function


### Fourier Series Approximation 

```python

import plotly.graph_objects as go
import numpy as np
import pandas as pd

fourier=pd.DataFrame() # To collect data

x = np.linspace(0,2*np.pi,400) #Genearate 400 points between 0 and 2pi

a=np.empty(int(len(x)/2))
a.fill(1)
b=np.empty(int(len(x)/2)) # Step Function Declaration, first half will be equal to one, and second half will be equal to -1
b.fill(-1)
CC=np.append(a,b) #joins two halves


fourier['x']=x
fourier['actualData']=CC


nval=[20, 40, 80, 160, 500, 1000]#Different values of n for coefficients
for n in nval:
    L=np.pi
    a0=0 
    an=0  # Based on calculation, an and a0 is equal to 0
    bn=[]
    fx=[]
    for i in range(1,n+1):
        b=(1/(i*np.pi))*(2*(1-np.cos(i*np.pi))) #Generate bn coefficients based on formula
        #b = b*(1*i*(np.pi/(n/2)))
        bn.append(b)
    for a in x:
        result=0
        for i in range(1,n+1):
            result=result+(bn[i-1])*np.sin(i*a) #Calculate Fourier Series approximation using formula
        fx.append(result/1)# For some reason, I had to divide by 10 because that was where the fourier series was converging to, not sure why exactly.
    fourier['n=%s'%(n)]=fx #Add to data frame based on the value of n



```

```python
fourier #result
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['actualData'],mode='lines',name='+1-1Function'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=20'],mode='lines',name='n=20'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=40'],mode='lines',name='n=40'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=80'],mode='lines',name='n=80'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=160'],mode='lines',name='n=160'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=500'],mode='lines',name='n=500'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=1000'],mode='lines',name='n=1000'))
fig.show()
```

### Graphing error based on different n values

```python
error=pd.DataFrame() #Create new dataframe for error
for n in nval:
    error['n=%s'%(n)]=np.log10(np.abs(fourier['n=%s'%(n)]-fourier['actualData'])) #Compute the error by subtraction approximation with actual
```

```python
error
fig=go.Figure()
fig.add_trace(go.Scatter(x=fourier['x'],y=error['n=20'],mode='lines',name='n=20'))
fig.add_trace(go.Scatter(x=fourier['x'],y=error['n=40'],mode='lines',name='n=40'))
fig.add_trace(go.Scatter(x=fourier['x'],y=error['n=80'],mode='lines',name='n=80'))
fig.add_trace(go.Scatter(x=fourier['x'],y=error['n=160'],mode='lines',name='n=160'))
fig.add_trace(go.Scatter(x=fourier['x'],y=error['n=500'],mode='lines',name='n=500'))
fig.add_trace(go.Scatter(x=fourier['x'],y=error['n=1000'],mode='lines',name='n=1000'))
fig.show()
```
