### Fourier Series Approximation using Sin and Cos

```python
#ToDo: Numerically computation of Fourier Coefficients
import plotly.graph_objects as go
import numpy as np
import pandas as pd

fourier=pd.DataFrame() # To collect data
fourierCoefficients=pd.DataFrame()
L=np.pi
X = np.linspace(-L,L,400) #Genearate 400 points between 0 and 2pi
alpha=6
beta=5
actual=[]
def f(x):
    fx=np.sin(alpha*x)+np.sin(beta*x)
    return fx
for x in X:
    actual.append(f(x))
nval=[20, 40, 80, 160, 500, 1000]#Different values of n for coefficients
for N in nval:
    
    a0=0
    an=[]  # Based on calculation, an and a0 is equal to 0
    bn=[]
    fx=[]
    for n in range(1,N+1):
        a=0
        an.append(a)
        if(n==alpha or n==beta):
            bn.append(1)
        else:
            k=(2*n*np.sin(np.pi*alpha)*np.cos(np.pi*n))-(2*alpha*np.cos(np.pi*alpha)*np.sin(np.pi*n))
            j=(2*n*np.sin(np.pi*beta)*np.cos(np.pi*n))-(2*beta*np.cos(np.pi*beta)*np.sin(np.pi*n))
            b=(k/((np.pi*alpha**2)-(np.pi*n**2)))+(j/((np.pi*beta**2)-(np.pi*n**2)))
            bn.append(float(b))
    for a in X:
        result=a0
        for i in range(1,N+1):
            result=result+(bn[i-1])*np.sin(i*a) #Calculate Fourier Series approximation using formula
        fx.append(result)# For some reason, I had to divide by 10 because that was where the fourier series was converging to, not sure why exactly.
    fourier['n=%s'%(n)]=fx #Add to data frame based on the value of n
    

fourier['x']=X
fourier['actualData']=actual

fig=go.Figure()
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['actualData'],mode='lines',name='+1-1Function'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=20'],mode='lines',name='n=20'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=40'],mode='lines',name='n=40'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=80'],mode='lines',name='n=80'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=160'],mode='lines',name='n=160'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=500'],mode='lines',name='n=500'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=1000'],mode='lines',name='n=1000'))

fig.show()
display(fourier.head(15))
```

### Fourier Series Approximation using Complex number

```python

import plotly.graph_objects as go
import numpy as np
import pandas as pd
fourier=pd.DataFrame() # To collect data
nval=[20, 40, 80, 160, 500, 1000]#Different values of n for coefficients

L=np.pi
X = np.linspace(-L,L,400) #Genearate 400 points between 0 and 2pi
a=6
b=5
actual=[]
def f(x):
    fx=np.sin(a*x)+np.sin(b*x)
    return fx
for x in X:
    actual.append(f(x))

for N in nval:
    cn=[]
    fx=[]
    for n in range(int((-N/2)),int((N/2)+1)):
            if n==a or n==b or n==-a or n==-b:
                cn.append(1)
            else:
                c=(1/(2*np.pi))*2*1j*((a*np.cos(np.pi *a)* np.sin(np.pi*n)-n*np.sin(np.pi*a)*np.cos(np.pi*n))/(a**2-n**2)+((b*np.cos(np.pi*b)*np.sin(np.pi*n) - n*np.sin(np.pi*b)*np.cos(np.pi*n))/(b**2 - n**2)))
                cn.append(c) 
    
    for x in X:
        result=0
        for i in range(int((-N/2)),int((N/2))):
            result=result+cn[int(i+(N/2))]*(np.e**(1j*i*x)) #Calculate Fourier Series approximation using formula
        fx.append(result.real)
    fourier['n=%s'%(N)]=fx #Add to data frame based on the value of n()
fourier['x']=X
fourier['actualData']=actual
fig=go.Figure()
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['actualData'],mode='lines',name='+1-1Function'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=20'],mode='lines',name='n=20'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=40'],mode='lines',name='n=40'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=80'],mode='lines',name='n=80'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=160'],mode='lines',name='n=160'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=500'],mode='lines',name='n=500'))
fig.add_trace(go.Scatter(x=fourier['x'],y=fourier['n=1000'],mode='lines',name='n=1000'))
fig.show()
display(fourier.head(30)) #Shows the first 30 entries of the table
```
