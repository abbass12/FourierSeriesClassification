Cn approximation for sin(alpha(x)+sin(beta*x)



```python
import numpy as np
import plotly.graph_objects as go


M=100 #number of gir
alpha=5
beta=6
cn=[]
fx=[]

def f(x):
    #return np.sin(alpha*x)+np.sin(beta*x)
    return np.exp(-25*(x/(np.pi/4))**20)
def e(n,x):
    return np.e**(-1j*n*x)
def e2(x,n):
    return np.e**(1j*x*n)    
n=np.linspace(-M/2,M/2,M-1)
h=np.linspace(0,M-1,M-1) # 
xi=-np.pi+(2*np.pi/M)*h

M2 = 1500
h2=np.linspace(0,M2-1,M2-1) #
xi2 =-np.pi+ (2*np.pi/M2)*h2
Fx=f(xi)

D=np.zeros((M-1,M-1),dtype=complex)
for p in range(M-1):
  for q in range(M-1):
    D[p][q]=(e(n[p],xi[q]))
f=f(xi)
cn=np.dot(D,f)/M
#cn=(cn.sum(axis=1))/M

# Computing edges
sig=np.zeros((M-1),dtype=complex)
sig = np.pi*(np.abs(n)/(M/2))
cn = cn*(1j*np.sign(n)*sig)

D2=np.zeros((M2-1,M-1),dtype=complex)
for p in range(M2-1):
  for q in range(M-1):
    D2[p][q]=(e2(xi2[p],n[q]))

fx2 = (np.dot(D2,cn)).real
#fx=(cn*(np.e**(1j*n*xi))).real

fig=go.Figure()
fig.add_trace(go.Scatter(x=xi,y=Fx,mode='lines',name='+Actual Function-'))
fig.add_trace(go.Scatter(x=xi2,y=fx2,mode='lines',name='+Fourier Approximation-'))

fig.show()
```
